```python
#tokenization
from nltk.tokenize import word_tokenize
words=word_tokenize(text)
words
```

```python
 #frequency of words
from nltk.probability import FreqDist
fdist=FreqDist()
for word in words:
    fdist[word.lower()]+=1
fdist
list(fdist)
```

```python

# how many time he uses word "attack" and "terrorist" 
print(fdist["threats"],"&",fdist["terrorism"],"Times Respectively ")
```

```python

# how many different word Dr. S jaishakar used in his speech
len(fdist)
```

```python
# most common word he used 
print(fdist.most_common(10))
```

```python

# Stop words : have no meaning in analytics
from nltk.corpus import stopwords
st=stopwords.words('english')
st
```

```python

# remove the stop words from speech and look about what the actully speech was
all_words=list(fdist)
means=[]
for i in all_words:
    if i in st:
        pass
    else:
        means.append(i)
means
```

```python

# conclusion : Dr. S Jaishankar talked about : 
for i in range(10):
    if len(means[i])>3:
        print(means[i])
```
